### Transformar las variables al formato correcto

```python
df_temp['year'] = pd.to_datetime(df_temp['dt'])
df_temp['year']
```

### Crea un filtro de las fechas mayores que necesitas analizar y la guardas en el mismo df

```python
idx = df_temp['year'] > pd.to_datetime('1970-01-01')
df_temp = df_temp[idx]
df_temp
```

```python
df_t_avg = df_temp.groupby(['Country',
           pd.Grouper(key='year',freq='1Y')]).agg({'AverageTemperature':[np.mean,np.median]})
```

```python
df_t_avg
```

## La siguiente función filtra un indice y un valor especifico para graficar 

```python
df_t_avg.xs(['Mexico'])['AverageTemperature'].plot()
```

```python
df_t_med = df_t_avg['AverageTemperature'][['median']].reset_index()
```

## Extraer de la fecha solamente el año.

```python
df_t_med['date'] = df_t_med['year'].dt.year
df_t_med['date']
```

## Renombrar variables

```python
df_t_med.rename(columns={'median':'temperature'},inplace=True)
df_t_med
```

## La función pivot para crear una tabla dinamica desde tu dataframe, nos da una ventaja para modelar los datos adecuadamente y graficar casi cualquier cosa

Asombrosamente existe esta forma de reestructurar los datos para poder graficar ó aplicar diferentes graficas exploratorias que nos dan un punto de vista mejor de la información que contiene nuestro data frame.

```python
df_t_pivot = df_t_med.pivot_table(values='temperature',index='date',columns='Country')
df_t_pivot
```

#### Lo primero que hicimos fue que si quieres comparar dos paises, obviamente tus dos paises deben ser columnas ó tambien conocidas como variables, el error mio anterior era que estaba usando a los paises como indices de primer nivel y no como variables. En este caso, cada observación de cada variable me permitirá aplicarle un loc a las columnas mostrando todos los datos para poder comparar el comportamiento de los mismos en cada variable.

```python
df_t_pivot.loc[:,['Afghanistan','Colombia']].plot()
```

### Basicamente cuando defines un index, estas definiendo el eje x de tus datos, los datos que estarán en función de las variables. 

```python
df_t_pivot.boxplot()
```

### Importante manejar la transpuesta de tus datos, te sirve mucho para manejar funciones y si las quieres aplicar a tus columnas o a tus filas, la transpuesta te sirve si los datos que quieres manejar se encuentran en tus columnas y quieres hacer por ejemplo un sample, en este caso se aplica la función a la transpuesta, y imagino devolvemos con la transpuesta de la transpuesta los datos a su origen

```python
df_t_pivot.T.sample(10).T.boxplot(rot=45)
```

```python
df_agri = pd.read_csv('API_AG.LND.AGRI.K2_DS2_en_csv_v2_716226.csv', header = 2)
df_fore = pd.read_csv('API_AG.LND.FRST.K2_DS2_en_csv_v2_716262.csv', header = 2)
df_elec = pd.read_csv('API_EG.USE.ELEC.KH.PC_DS2_en_csv_v2_715482.csv', header = 2)
df_co2e = pd.read_csv('API_EN.ATM.CO2E.KT_DS2_en_csv_v2_713263.csv', header = 2)
df_popu = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_713131.csv', header = 2)
```

```python
map(str,range(1971,2015))
cols= ['Country Name','Country Code'] + list(map(str,range(1971,2015)))
```

```python
df_agri.loc[:,cols].melt(id_vars=['Country Name','Country Code']).rename(columns ={'variable':'date','Country Name':'Country',
        'Country Code':'name',
        'value':'agriculture'})
```

```python
def fun_format(df,col='agriculture'):
    return df.loc[:,cols].melt(id_vars=['Country Name','Country Code']).rename(
        columns ={'variable':'date',
                  'Country Name':'Country',
                  'Country Code':'name',
                  'value':col})
```

```python
df_agri = fun_format(df_agri,col='agriculture')
df_fore = fun_format(df_fore,col='forest')
df_elec = fun_format(df_elec,col='electricprod')
df_co2e = fun_format(df_co2e,col='co2')
df_popu = fun_format(df_popu,col='population')
```

```python
df_popu['date'] = df_popu['date'].astype(float)
df_fore['date'] = df_fore['date'].astype(float)
df_agri['date'] = df_agri['date'].astype(float)
df_elec['date'] = df_elec['date'].astype(float)
df_co2e['date'] = df_co2e['date'].astype(float)
```

```python
df_merge = pd.merge(df_t_med[['Country','temperature','date']],
                   df_popu, on = ['Country','date'],how='inner')
```

```python
df_merge
```

```python
df_merge = pd.merge(df_merge,
                    df_co2e,
                    on = ['Country','name','date'],
                    how = 'inner')

df_merge = pd.merge(df_merge,
                    df_fore,
                    on = ['Country','name','date'],
                    how = 'inner')

df_merge = pd.merge(df_merge,
                    df_elec,
                    on = ['Country','name','date'],
                    how = 'inner')

df_merge = pd.merge(df_merge,
                    df_agri,
                    on = ['Country','name','date'],
                    how = 'inner')
```

```python
df_merge = pd.merge(df_merge,
                    df_agri,
                    on = ['Country','name','date'],
                    how = 'inner')
```
