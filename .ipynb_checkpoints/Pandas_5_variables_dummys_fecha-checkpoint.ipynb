## Copia de Bases

Importante siempre colocar que la copia va a ser profunda, ya que si no colocas la copia profunda los datos que se trasladan a df van a ser indices hacia la memoria donde esta guardada la base original, por tanto, si tu modificas la base original tu supuesta copia tambien va a sufrir cambios. 

```python
df = df_development.copy(deep=True)
df.drop(['id'],axis = 1, inplace = True)
```
