```python
def mean_eur2usd(x):
    return np.mean(x)*1.12
```

### ¿Como aplicar funciones a la agrupación, tambien con apply?

```python
df.groupby('sex')[['prct_tip','total_bill']].aggregate([mean_eur2usd,np.mean])
```

```python
df.groupby('sex')['prct_tip'].aggregate([mean_eur2usd,np.mean])
```

```python
df.groupby(['sex','time'])[['prct_tip','total_bill']].aggregate([mean_eur2usd,np.mean])
```

### Tambien permite aplicar funciones definidas dentro de la libreria numpy

### Tambien puedo aplicar la función interna de lambda para calcularlo

```python
df.groupby(['sex','time'])[['prct_tip','total_bill']].apply(lambda x: np.mean(x)*1.12)
```

### La función aggregate nos permite agregar mas de una función dentro de la consulta

```python
dict_agg = {'tip': [min,max], 'total_bill':[np.mean,mean_eur2usd]}
```

```python
df.groupby(['sex','time'])[['tip','total_bill']].agg(dict_agg)
```

### Crear una función que retorne valores gracias a una validación logica interna, esta función puede ser incluida al momento de aplicar el groupby  gracias a la extención de función que tiene .filter()

```python
def f_filter(x):
    return mean_eur2usd(x['total_bill'].mean()) > 20
```

```python
df_filtered = df.groupby(['sex','time']).filter(f_filter)
```

```python
df_filtered.groupby(['sex','time']).count()
```
