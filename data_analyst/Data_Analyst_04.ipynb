```python
import scipy.stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('bicicletas-compartidas.csv')
```

## Categoricas

```python
y = df['cuartil-ano'].apply(lambda x: 'cuartil-' + str(int(x)))
```

```python
fig,axes = plt.subplots()
axes.bar(y.value_counts().index,y.value_counts())
axes.set_xlabel('Cuartiles del a√±o')
axes.set_ylabel('Frecuencia')
```

```python
fig,axes = plt.subplots()
axes.bar(y.value_counts().index,y.value_counts())
axes.set_xlabel('Cuartiles del a√±o')
axes.set_ylabel('Frecuencia')
axes.patches[3].set_facecolor('red') # Llamar los patches es los valores que toma la variable
```

```python
fig,ax = plt.subplots()
ax.pie(y.value_counts(),labels = y.value_counts().index)
ax.set_title('Diagrama de pie')
```

## Numericas

```python
y = df['viento']
```

```python
fig,ax = plt.subplots()
ax.hist(y,bins = 30)
ax.set_xlabel('Viento')
ax.set_ylabel('Frecuencias')
plt.axvline(np.mean(y),c = 'r',linestyle = '--',label = 'Prom')
plt.axvline(np.mean(y) + np.std(y),c='k',linestyle='--',label = 'std + 1')
plt.axvline(np.mean(y) - np.std(y),c='k',linestyle='--',label = 'std + 1')
plt.legend()
plt.show()
```
