```python
import numpy as np
from matplotlib import pyplot 
from numpy.random import normal
from scipy.stats import norm
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\GAMER\\Documents\\Development\\curso_python\\base\\avocado.csv')
```

```python
df_avocado = df[df.notnull()]
```

## Estimaci√≥n Parametrica

```python
mu = df_avocado['AveragePrice'].mean()
sigma = df_avocado['AveragePrice'].std()
dist = norm(mu,sigma)
values = np.arange(0.2,3.3,0.1) # Modificaci√≥n a la escala
probabilities = [dist.pdf(value) for value in values]
pyplot.hist(df_avocado['AveragePrice'], bins=30, density=True)
pyplot.plot(values, probabilities)
pyplot.show()
```

## Estimaci√≥n No Parametrica

```python
from numpy import hstack
from sklearn.neighbors import KernelDensity

model = KernelDensity(bandwidth=0.07,kernel='gaussian')
sample = df_avocado['AveragePrice']
sample = sample.to_numpy()
sample = sample.reshape(len(sample),1)
model.fit(sample)
```

```python
values = np.arange(0.2,3.3,0.1)
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values) #probabilidad logar√≠tmica
probabilities = np.exp(probabilities)  # inversi√≥n de probabilidad

pyplot.hist(sample, bins=30, density=True) 
pyplot.plot(values[:], probabilities)
pyplot.show()
```
